# II. Exploring and Visualizing Big Five Personality Dimensions


A brief examination of the TIPI scores reveals that they represent the Big Five personality traits (openness, conscientiousness, extraversion, agreeableness, and neuroticism), with each item representing a polar end of a trait dimension. TIPI responses can be manipulated to return Big Five scores for every respondent, and can subsequently lead to extrapolation at a demographic level.

Because these items were scored on a likert scale, we will need to reverse certain items before combining with their counterpart. Once this is completed, we will create a feature for each Big Five trait.

```python
DASSdata['extraversion'] = DASSdata['TIPI1'] + (7 - DASSdata['TIPI6'])
DASSdata['agreeableness'] = DASSdata['TIPI2'] + (7 - DASSdata['TIPI7']) 
DASSdata['conscientiousness'] = DASSdata['TIPI3'] + (7 - DASSdata['TIPI8']) 
DASSdata['neuroticism'] = DASSdata['TIPI4'] + (7 - DASSdata['TIPI9'])
DASSdata['openness'] = DASSdata['TIPI5'] + (7 - DASSdata['TIPI10']) 
```

Excellent, now let's get an idea of the distribution for each trait.

```python
# I have 5 plots that i want to show in 2 rows. So I do 3 columns. That way i have 6 plots.
f, axes = plt.subplots(2, 3, figsize=(15, 10))

sns.countplot(x='extraversion', palette='BrBG', ax=axes[0, 0], data=DASSdata)
sns.countplot(x='agreeableness', palette='PiYG', ax=axes[0, 1], data=DASSdata)
sns.countplot(x='conscientiousness', palette='PuOr', ax=axes[0, 2], data=DASSdata)
sns.countplot(x='neuroticism', palette='PRGn', ax=axes[1, 0], data=DASSdata)
sns.countplot(x='openness', palette='RdBu', ax=axes[1, 1], data=DASSdata)

# If you have a sixth plot, you can add it here, or remove the empty subplot
# For example:
# sns.countplot(x='your_sixth_column', palette='YourPalette', ax=axes[1, 2], data=DASSdata)

# Hide the empty subplot (if there is one) like this:
axes[1, 2].set_visible(False)  # This hides the bottom right plot

plt.tight_layout()
plt.show()
```

This sets a good basis for our Big Five profile. We could spend a considerable amount of time investigating how each of these traits interacts with our other variables, but it this is not within the scope of our project. Instead, we will settle on highlighting a few select areas - namely, whether there is a dependent relationship between Big Five traits and religion, orientation, and level of education. We will utilize the chi-squared test of independence and work with the following general hypotheses:

Null hypothesis (H0): The given variable (religion, orientation, or level of education) is independent of the given Big Five trait.
Alternative hypothesis (Ha): The given variable (religion, orientation, or level of education) is dependent on the given Big Five trait.

# Religion

```python
import scipy
from scipy import stats

#first we'll construct a contingency table for each trait
tab_religion_O =  pd.crosstab(DASSdata['religion'], DASSdata['openness'])
tab_religion_C =  pd.crosstab(DASSdata['religion'], DASSdata['conscientiousness'])
tab_religion_E =  pd.crosstab(DASSdata['religion'], DASSdata['extraversion'])
tab_religion_A =  pd.crosstab(DASSdata['religion'], DASSdata['agreeableness'])
tab_religion_N =  pd.crosstab(DASSdata['religion'], DASSdata['neuroticism'])
```

```python
#building a function to neatly print each test's results
def results(chi_squared, p, dof, expected):
    print(f"Chi-Squared: {chi_squared}")
    print(f"P-value: {p}")
    print(f"Degrees of Freedom: {dof}") 
```

```python
#and now we can run our tests
chi_squared, p, dof, expected = stats.chi2_contingency(tab_religion_O)
print("Results of Big Five - Openness x Religion:")
results(chi_squared, p, dof, expected)
```

```python
chi_squared, p, dof, expected = stats.chi2_contingency(tab_religion_C)
print("Results of Big Five - Conscientiousness x Religion:")
results(chi_squared, p, dof, expected)
```

```python
chi_squared, p, dof, expected = stats.chi2_contingency(tab_religion_E)
print("Results of Big Five - Extraversion x Religion:")
results(chi_squared, p, dof, expected)
```

```python
chi_squared, p, dof, expected = stats.chi2_contingency(tab_religion_C)
print("Results of Big Five - Agreeableness x Religion:")
results(chi_squared, p, dof, expected)
```

```python
chi_squared, p, dof, expected = stats.chi2_contingency(tab_religion_N)
print("Results of Big Five - Neuroticism x Religion:")
results(chi_squared, p, dof, expected)
```

We can ascertain from these tests that it is safe to reject all nulls and conclude that religion and each of the Big Five traits are dependent.

In this subsection, we'll be using the random forest model, an ensemble method which consists of the aggregation of many decision trees. We'll start by considering how we approach our model settings.

Hyperparameter Tuning and Model Fitting
With each model comes a considerable variety of hyperparameters, or settings, that influence the model's scoring process. These hyperparameters often benefit from customization, with the downside of suffering from tedious, but necessary re-fitting of the model for each parameter change.

A more succinct method of comparing hyperparameter choices is to construct a validation curve, which plots a given range for the selected hyperparameter, allowing for the model-builder to more easily select the most effective option. In this case, the best option is the point at which the cross-validation (CV) score begins to plateau.

Let's demonstrate this process:

```python
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.model_selection import ValidationCurve

# Assuming X and y are already defined in your workspace as your features and target variable respectively

viz = ValidationCurve(
    RandomForestClassifier(), param_name="max_depth",
    param_range=np.arange(1, 10), cv=5, scoring="r2"
)


viz.fit(X, y)
viz.show()
```

Re-evaluting Our Approach
There is one possible line of inquiry left to us, one which will require a shift in perspective.

We'll go back to the beginning of our initial machine learning purpose. Our goal was to predict responses to Q13 on a 1-4 scale. Let's return to this point for a second: how do we weigh the costs vs. benefits of predicting a given class? In the classic example of predicting cancer diagnosis, there is a considerably heavier cost attributed to predicting a false negative as opposed to a false positive; it is far worse to tell someone who has cancer that they don't have the disease than it is to diagnose someone as having cancer when they are actually cancer-free.

We can apply a similar line of thought here with Q13: are weights across our four predicted classes entirely equal , are the separations between classes somethng that ought to be maintained? In other words, is it important to be able to differentiate 4 vs. 3 in a given instance? Or are we merely concerned with distinguishing between those instances that are "less likely" or "more likely" to experience feelings of sadness/depression?

If this latter statement is accepted as sufficient, then we can reformat our dataset into a binary classification, thereby simplifying the prediction process. We will group 1 and 2 into the "less likely" class (0), then group 3 and 4 into a "more likely" class (1), and rerun our random forest classifier.

```python
#regrouping
DASSdata.loc[DASSdata['Q13A'] <= 2, 'Q13A'] = 0 
DASSdata.loc[DASSdata['Q13A'] >= 3, 'Q13A'] = 1
```

```python
#refitting
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(max_depth = 9, n_estimators = 150, random_state = 0)

RFC.fit(X2_train, y2_train)
```

```python
RFC.score(X2_test, y2_test)
```

```python
from yellowbrick.classifier import classification_report

# Instantiate the visualizer
visualizer = classification_report(
    RandomForestClassifier(), X2_train, y2_train, X2_test, y2_test, support=True)
```
